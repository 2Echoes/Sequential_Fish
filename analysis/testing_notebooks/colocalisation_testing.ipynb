```python
import pandas as pd
import numpy as np
PATH = "/media/floricslimani/SSD4To/SSD_floricslimani/Fish_seq/Davide/2024-08-12 - SeqFISH - HeLa - Puro - R2TP1-2_Run7/"
```

```python
Acquisition = pd.read_feather(PATH + "result_tables/Acquisition.feather")
Detection = pd.read_feather(PATH + "result_tables/Detection.feather")
Colocalisation = pd.read_feather(PATH + "result_tables/Colocalisation.feather")
Gene_map = pd.read_feather(PATH + "result_tables/Gene_map.feather")
```

# Preprocessing

## Merging data

```python
coloc_len = len(Colocalisation)

Detection = pd.merge(
    Detection,
    Acquisition.loc[:,['acquisition_id','cycle']],
    on='acquisition_id'
)

Detection = pd.merge(
    Detection,
    Gene_map.loc[:,['target','cycle','color_id']],
    on=['color_id', 'cycle']
)
Colocalisation = pd.merge(
    Colocalisation,
    Detection.loc[:,['target','detection_id']],
    left_on='detection_id1',
    right_on='detection_id'
).rename(columns={'target' : 'target1'}).drop(columns= 'detection_id')
Colocalisation = pd.merge(
    Colocalisation,
    Detection.loc[:,['target','detection_id']],
    left_on='detection_id2',
    right_on='detection_id'
).rename(columns={'target' : 'target2'}).drop(columns= 'detection_id')
 
assert len(Colocalisation) == coloc_len
Colocalisation
```
